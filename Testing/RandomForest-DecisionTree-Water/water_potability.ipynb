```python
#Variables predictoras
x_frame = dfWater2.drop(['Potability'], axis=1)
X = np.array(x_frame)
pd.DataFrame(X)
```

```python
# Variable Clase:
y_frame = dfWater2['Potability']
Y = np.array(y_frame)
pd.DataFrame(Y)
```

```python
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, 
                                                                                test_size = 0.2, 
                                                                                random_state = 0,
                                                                                shuffle = True)
```

```python
print(len(X_train))
print(len(X_validation))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Se entrena el modelo a partir de los datos de entrada
ClasificacionAD = DecisionTreeClassifier(random_state=0)
ClasificacionAD.fit(X_train, Y_train)

# ClasificacionAD = DecisionTreeClassifier(max_depth=14, 
#                                         min_samples_split=4, 
#                                         min_samples_leaf=2,
#                                         random_state=0)
# ClasificacionAD.fit(X_train, Y_train)
```

```python
#Clasificación final 
Y_ClasificacionAD = ClasificacionAD.predict(X_validation)
#print(Y_ClasificacionAD)
```

```python
ValoresMod1 = pd.DataFrame(Y_validation, Y_ClasificacionAD)
ValoresMod1
```

```python
#Matriz de clasificación
ModeloClasificacion1 = ClasificacionAD.predict(X_validation)
Matriz_Clasificacion1 = pd.crosstab(Y_validation.ravel(), 
                                    ModeloClasificacion1, 
                                    rownames=['Reales'], 
                                    colnames=['Clasificación']) 
Matriz_Clasificacion1
```

```python
#Reporte de la clasificación
print(classification_report(Y_validation, Y_ClasificacionAD))
print('Criterio: \n', ClasificacionAD.criterion)
print('Importancia variables: \n', ClasificacionAD.feature_importances_)
print("Exactitud: "+str(accuracy_score(Y_validation, Y_ClasificacionAD)*100)+"%")
precisionAD = float(classification_report(Y_validation, Y_ClasificacionAD).split()[10])*100
print("Precisión: "+str(precisionAD)+"%")
print("Tasa de error: "+str((1-ClasificacionAD.score(X_validation, Y_validation))*100)+"%")
sensibilidadAD = float(classification_report(Y_validation, Y_ClasificacionAD).split()[11])*100
print("Sensibilidad: "+str(sensibilidadAD)+"%")
especificidadAD = float(classification_report(Y_validation, Y_ClasificacionAD).split()[6])*100
print("Especificidad: "+str(especificidadAD)+"%")
```

```python
# Graficamos la curva ROC con Plotly
y_score1 = ClasificacionAD.predict_proba(X_validation)[:,1]
y_score2 = ClasificacionBA.predict_proba(X_validation)[:,1]

import plotly.express as px
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(Y_validation2, y_score1)
fpr2, tpr2, thresholds2 = roc_curve(Y_validation2, y_score2)

# Graficamos la curva ROC con Plotly
fig = px.area(title='Curva ROC Árbol de decisión vs Bosque aleatorio')
# Agregamos la curva ROC del arbol de decisión más el ROC de este modelo y agregamos un área sombreada debajo de la curva
fig.add_scatter(x=fpr, y=tpr, mode='lines', name='Árbol de decisión, AUC = '+str(auc(fpr, tpr).round(4)), fill='tonexty')
# Agregamos la curva ROC del bosque aleatorio más el ROC de este modelo
fig.add_scatter(x=fpr2, y=tpr2, mode='lines', name='Bosque aleatorio, AUC = '+str(auc(fpr2, tpr2).round(4)), fill='tonexty')
# Agregamos la diagonal de la curva ROC con líneas punteadas
fig.add_shape(type="line", x0=0, y0=0, x1=1, y1=1, line=dict(color="Black", dash="dash"))

fig.update_layout(yaxis_title='True Positive Rate', xaxis_title='False Positive Rate')
fig.show()
```
